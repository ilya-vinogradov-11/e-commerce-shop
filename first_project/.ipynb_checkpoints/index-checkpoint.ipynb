```python
import pandas as pd
import requests 
from urllib.parse import urlencode 
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
def parse_link(link):
    base_url = 'https://cloud-api.yandex.net/v1/disk/public/resources/download?'
    final_url = base_url + urlencode(dict(public_key=link))
    response = requests.get(final_url)
    download_url = response.json()['href']
    download_response = requests.get(download_url) 
    return download_url
```

```python
olist_customers_dataset  = pd.read_csv(parse_link('https://disk.yandex.ru/d/aFh0UoU0BrkIRg'))
```

```python
olist_orders_dataset = pd.read_csv(parse_link('https://disk.yandex.ru/d/hgBCExTG47kxlQ'),parse_dates=\
                       ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',\
                        'order_delivered_customer_date', 'order_estimated_delivery_date'])
```

```python
olist_order_items_dataset = pd.read_csv(parse_link('https://disk.yandex.ru/d/Cu2UJxIUUfltVA'))
```

# ПЯТЫЙ ШАГ: 
## провести когортный анализ пользователей. В период с января по декабрь выявить когорту с самым высоким retention на 3й месяц.

#### В данном шаге я разбиваю пользователей на когорты, ориентируясь на их первую покупку и месяц, в который она была совершена.

```python
five_step_df = olist_orders_dataset_with_approved_orders.copy() # берем данные подтвержденных покупок

months_list = [
    'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'
]
```

```python
five_step_df["customer_unique_id"] = five_step_df["customer_id"]\
                                                            .map(olist_customers_dataset\
                                                            .set_index("customer_id")["customer_unique_id"])
# добавляем уникальные айди пользователей из соответствующего датафрейма
```

```python
five_step_df = five_step_df.sort_values(by=['customer_unique_id', 'order_approved_at' ])
```

```python
first_purchase_per_user = five_step_df.drop_duplicates(subset='customer_unique_id', keep='first')

# создаем датафрейм с первыми покупками...
```

```python
first_purchase_per_user = first_purchase_per_user.copy()
first_purchase_per_user['cohort'] = first_purchase_per_user['order_approved_at'].dt.strftime('%B')
# ...создаем колонку с принадлежностью к определенной когорте...
```

```python
five_step_df["cohort"] = five_step_df["customer_unique_id"]\
                         .map(first_purchase_per_user\
                         .set_index("customer_unique_id")["cohort"])
# ...и добавляем ее в изначальный датафрейм
```

```python
five_step_df['third_month'] = five_step_df['cohort']
# создаем колонку с третьим месяцем для каждой когорты. В будущем мы ее изменим.
five_step_df['cohort'] = 'cohort_' + five_step_df['cohort']
# изменяем значения когорт чтоб не путаться 
```

```python
def find_third_month(month):
    idx = months_list.index(month)
    next_idx = (idx + 3) % len(months_list)  
    return months_list[next_idx] 
```


#### данная функция подбирает для каждой когорты соответствующий третий месяц. 
#### Например для Января(01), таким месяцем будет:01 + 1 + 1 + 1 = 04 - Апрель


```python
five_step_df['third_month'] = five_step_df['third_month'].apply(find_third_month) 
```

```python
highests_cogorts = five_step_df.query('third_month == order_approved_at.dt.strftime("%B")')\
                  .groupby(['cohort', 'third_month'])\
                  .agg({'order_approved_at': 'count'}).reset_index().sort_values(by='order_approved_at')
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.barplot(x='third_month', y='order_approved_at', hue='cohort', data=highests_cogorts)
plt.title('Количество покупок когорт на третий месяц')
plt.xlabel('Месяц')
plt.ylabel('Количество покупок')
plt.legend(title='Когорта', bbox_to_anchor=(1, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

### ВЫВОД: как можно видеть из графика, наибольший retention на третий месяц продемонстрировала мартовская когорта. В ее случае третьим месяцем является Июнь.

# ШЕСТОЙ ШАГ: 
## Используя python, построить RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию

```python
six_step_df = olist_orders_dataset_with_approved_orders.copy() # берем данные подтвержденных покупок
```

```python
six_step_df["customer_unique_id"] = six_step_df["customer_id"]\
                                                            .map(olist_customers_dataset\
                                                            .set_index("customer_id")["customer_unique_id"])
```

```python
six_step_df = six_step_df.sort_values(by=['customer_unique_id', 'order_approved_at' ])                                     
```

```python
last_purchase_per_user = six_step_df.drop_duplicates(subset='customer_unique_id', keep='last')
last_purchase_per_user = last_purchase_per_user.copy()
last_date = last_purchase_per_user['order_approved_at'].max() + dt.timedelta(days=1)
```

#### в качестве нынешней даты(в которой нахожусь я), я решил выбрать последнюю дату среди всех покупок + еще один день 

```python
last_purchase_per_user['days_from_date_of_purchase'] = last_purchase_per_user['order_approved_at']\
                                                       .apply(lambda x: (last_date - x).days)
```


#### нашли количество дней, прошедших со дня последней покупки, для каждого пользователя

```python
count_purchases = six_step_df.groupby('customer_unique_id').agg({'order_approved_at': 'count'})\
                                         .sort_values(by = 'order_approved_at').reset_index()\
                                         .rename(columns = {'order_approved_at': 'all_purchases'})
# считаем количество покупок каждым пользователем...
```

```python
last_purchase_per_user['all_purchases'] = last_purchase_per_user['customer_unique_id']\
                                                            .map(count_purchases\
                                                            .set_index('customer_unique_id')['all_purchases'])
# ...и добавляем колонку 
```

#### наши количество покупок для каждого пользователя

```python
sum_product_for_customer = olist_order_items_dataset.copy() # берем данные о заказах
approved_orders = olist_orders_dataset_with_approved_orders['order_id']
sum_product_for_customer = sum_product_for_customer.query('order_id in @approved_orders')
# оставляем нужные заказы
```

```python
sum_product_for_customer["customer_unique_id"] = sum_product_for_customer["order_id"]\
                                                            .map(olist_orders_dataset_with_approved_orders\
                                                            .set_index("order_id")["customer_unique_id"])
```

```python
sum_product_for_customer = sum_product_for_customer.groupby("customer_unique_id").agg({'price': 'sum'})\
                                                     .rename(columns={'price': 'sum_purchases'}).reset_index()
```

#### нашли сумму всех покупок для каждого пользователя

```python
last_purchase_per_user['sum_purchases'] = last_purchase_per_user['customer_unique_id']\
                                                        .map(sum_product_for_customer\
                                                        .set_index('customer_unique_id')['sum_purchases'])
# добавляем колонку в основной датафрейм
```

```python
rfm_df = last_purchase_per_user[['customer_unique_id', 'days_from_date_of_purchase', 'all_purchases', 'sum_purchases']]
```

```python
rfm_df.head()
```

### Таким образом мы создали датафрейм с нужными, для расчета RFM, колонками.

## Cоздание кластеров
### R - Recency (R):
1 - Самые свежие покупки - 30 или меньше дней\
2 - Покупки за последние 31-60 дней\
3 - Покупки за последние 61-120 дней\
4 - Покупки за последние 121-200 дней\
5 - Наиболее старые покупки - более 200 дней
### Frequency (F):
1 - Количество покупок больше 6\
2 - Средняя частота покупок - 3-6\
3 - Единичные покупки - 1-2
### Monetary (M):
1 - Высокая сумма покупок - свыше 1000 долларов\
2 - Значительная сумма покупок - 500-1000 долларов\
3 - Умеренная сумма покупок - 100-500 долларов\
4 - Небольшая сумма покупок - 30-100 долларов\
5 - Самые маленькие покупки - менее 30 долларов

```python
def count_recency(days):
    bins = [-float('inf'), 30, 60, 120, 200, float('inf')]
    clust_numbers = [1, 2, 3, 4, 5]
    return pd.cut(days, bins=bins, labels=clust_numbers)

def count_frequency(purchases):
    bins = [-float('inf'), 2, 6, float('inf')]
    clust_numbers = [1, 2, 3]
    return pd.cut(purchases, bins=bins, labels=clust_numbers)

def count_monetary(money):
    bins = [-float('inf'), 30, 100, 500, 1000, float('inf')]
    clust_numbers = [5, 4, 3, 2, 1]
    return pd.cut(money, bins=bins, labels=clust_numbers)
```

```python
rfm_df = rfm_df.copy()
rfm_df['recency']  = count_recency(rfm_df['days_from_date_of_purchase'].values)
rfm_df['frequency']  = count_frequency(rfm_df['all_purchases'].values)
rfm_df['monetary']  = count_monetary(rfm_df['sum_purchases'].values)
```

```python
rfm_df['clusters'] = rfm_df['recency'].astype(str) + rfm_df['frequency'].astype(str) + rfm_df['monetary'].astype(str)
```

```python
rfm_df = rfm_df.sort_values(by='days_from_date_of_purchase')
fig, ax = plt.subplots(figsize=(15, 6))
ax.scatter(x = rfm_df['clusters'], y = rfm_df['days_from_date_of_purchase'])
plt.xticks(rotation=45)
plt.title('Распределение дней, прошедших со дня последней покупки')
plt.xlabel("clusters")
plt.ylabel("days from last purchase")
```

```python
rfm_df = rfm_df.sort_values(by='all_purchases')
fig, ax = plt.subplots(figsize=(15, 6))
ax.scatter(x = rfm_df['clusters'], y = rfm_df['all_purchases'])
plt.xticks(rotation=45)
plt.title('Распределение количества покупок ')
plt.xlabel("Кластеры")
plt.ylabel("Количество покупок")
```

```python
rfm_df = rfm_df.sort_values(by='sum_purchases')
fig, ax = plt.subplots(figsize=(15, 6))
ax.scatter(x = rfm_df['clusters'], y = rfm_df['sum_purchases'])
plt.xticks(rotation=45)
plt.title('Распределение суммы денег, потраченных на покупки')
plt.xlabel("Кластеры")
plt.ylabel("Сумма покупок")
```
